# ğŸ“Œ Recursion in Python â€” Definition

Recursion ek aisa programming technique hai jisme function khud ko hi call karta hai problem solve karne ke liye.

â†’ Matlab function apne andar se again same function ko call kare, jab tak problem ka smallest part (base case) solve na ho jaye.

````python
# Factorial using Recursion

def factorial(n):
    if n == 1:                 # Base case
        return 1
    return n * factorial(n - 1)   # Recursive call

print(factorial(5))
````

# ğŸ“Œ try, except aur finally kya hote hain?

| Block     | Purpose                                                         |
| --------- | --------------------------------------------------------------- |

# Dunder Methods (ya Magic Methods / Special Methods)

Python me Dunder Methods (ya Magic Methods / Special Methods) wo methods hote hain jinke naam ke aage aur peeche double underscore __ lagta hai.

`ğŸ”¹ Dunder = Double + Under ( __ )`

Isliye inhe dunder methods bola jata hai.

### ğŸ“Œ Definition

Dunder Methods aise predefined special methods hote hain jo Python classes ke objects ke behavior ko control karne ke liye use hote hain.
Ye methods automatically call hote hain, hume manually call nahi karna padta.

| Method       | Purpose / Kaam                                                 |
| ------------ | -------------------------------------------------------------- |
| `__init__()` | Object initialize karne ke liye (Constructor)                  |
| `__str__()`  | Print formatting decide karta hai (`print(obj)`) par call hota |
| `__repr__()` | Object ka developer representation deta hai                    |
| `__len__()`  | `len(obj)` ka response return karta hai                        |
| `__add__()`  | `+` operator ka behavior define karta hai                      |
| `__sub__()`  | `-` operator ka behavior define karta hai                      |
| `__mul__()`  | `*` operator ka behavior define karta hai                      |
| `__eq__()`   | Comparison `==` ka behavior define karta hai                   |
| `__del__()`  | Object destroy hone par call hota (Destructor)                 |


````python
#Example

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"Name: {self.name}, Age: {self.age}"

p = Person("Bittu", 22)
print(p)
````

# ğŸ”¶ Getter

### ğŸ“Œ Getter kya hota hai?

Getter ek method hota hai jo private variable ki value ko read / access karne ke liye use hota hai.

Agar variable private hai (__variable) to usko class ke bahar direct access nahi kiya ja sakta.
Isliye Getter method ki help se value ko safely access kiya jata hai.

### Getter ka Rule

* Variable private hona chahiye (__name)
* Getter method value return karega (change nahi karega)

````python
class Person:
    def __init__(self, age):
        self.__age = age   # private variable

    def get_age(self):     # Getter method
        return self.__age

p = Person(22)
print(p.get_age())  # Getter ko call kiya
````

#### âš¡ Getter kab use hota hai?

* âœ” Jab aap private data ko read karna chahte ho
* âŒ Lekin usko change nahi karna chahte

# ğŸ”¶ Setter 

### ğŸ“Œ Setter kya hota hai?

Setter ek method hota hai jo private variable ki value ko update / change karne ke liye use hota hai.

Agar variable private hai (__variable) to class ke bahar se uski value direct change nahi ki ja sakti.
Isliye Setter method ki help se value ko safe tarike se modify kiya jata hai.

### ğŸ§  Setter ka Rule

* Variable private hona chahiye (__name)
* Setter method nayi value accept karega
* Setter method variable ko modify karega
* Validation (check) yahi par hoti hai
example: negative age, marks > 100, empty string etc.

````python
class Person:
    def __init__(self, age):
        self.__age = age    # private variable

    def set_age(self, new_age):   # Setter method
        if new_age > 0:
            self.__age = new_age
        else:
            print("Invalid age")
p = Person(22)

p.set_age(25)     # Age update hogi
p.set_age(-5)     # Invalid, setter value change nahi karega
````

### âš¡ Setter kab use hota hai?

* âœ” Jab private variable ki value change karni ho
* âœ” Jab aap update se pehle validation lagana chahte ho
* âŒ Direct assign karna allowed nahi hota

`Summery:-` Setter method private variable ki value ko modify (update) karne ke liye use hota hai.
Yeh method new value receive karta hai aur update karne se pehle validation bhi laga
sakta hai. Setter encapsulation provide karta hai kyunki yeh data ko secure aur
controlled tariqe se change karne deta hai.

# Python compile language hai ya interpreted?

`Python dono hai â€” Compiled bhi aur Interpreted bhi.`

##### âœ” Kyon Python Compiled bhi hai?
Jab hum .py file run karte hain to Python pehle code ko compile karke Bytecode (.pyc file) banata hai.

* Source Code (.py)  â†’  Bytecode (.pyc)

Ye step Compiler ka kaam hai
(isliye Python compiled language bhi mani jaati hai)

##### âœ” Kyon Python Interpreted bhi hai?

Jo bytecode banta hai wo direct CPU execute nahi kar sakta, is bytecode ko Python Virtual Machine (PVM) line by line interpret karke chalata hai.

`Bytecode (.pyc)  â†’  PVM (Interpreter)  â†’  Output`

#### ğŸ”¥ Interview Point (Must Mention)

Python bytecode me compile hoti hai, fir PVM us bytecode ko interpret karke execute karta hai.
That's why Python is both a compiled and interpreted language.

`Summery:-` Python na to pure compiled language hai na hi pure interpreted language.
Python source code pehle bytecode (.pyc) me compile hota hai.
Fir bytecode ko Python Virtual Machine (PVM) line-by-line interpret karke
execute karti hai. Isliye Python compiled + interpreted language dono hai.

# ğŸ§µ Python String

String ko single quotes ' ', double quotes " " ya triple quotes ''' ''' se banaya ja sakta hai.

````python
s1 = 'Hello'
s2 = "Python"
s3 = '''Welcome to Programming'''
````

````python
# 1ï¸âƒ£ upper() â†’ String ko uppercase me convert karta hai

text = "python"
print(text.upper())   # PYTHON

# 2ï¸âƒ£ lower() â†’ String ko lowercase me convert karta hai

text = "HELLO"
print(text.lower())   # hello

# 3ï¸âƒ£ title() â†’ Har word ka first letter uppercase

text = "welcome to python"
print(text.title())   # Welcome To Python

# 4ï¸âƒ£ capitalize() â†’ Sirf first character uppercase

text = "python programming"
print(text.capitalize())   # Python programming

# 5ï¸âƒ£ strip() â†’ Start/end spaces hata deta hai

text = "   hello   "
print(text.strip())   # hello

# 6ï¸âƒ£ replace(old, new) â†’ Words replace karta hai

text = "I love Java"
print(text.replace("Java", "Python"))   # I love Python

# 7ï¸âƒ£ split() â†’ String ko list me convert karta hai

text = "a-b-c-d"
print(text.split("-"))   # ['a', 'b', 'c', 'd']

# 8ï¸âƒ£ join() â†’ List ke items ko string me convert karta hai

list1 = ["a", "b", "c"]
print("-".join(list1))   # a-b-c

# 9ï¸âƒ£ find() â†’ Substring position return karta hai (milne par index, nahi mila to -1)

text = "Python is awesome"
print(text.find("is"))   # 7

# ğŸ”Ÿ count() â†’ Ek substring kitni baar aata hai

text = "banana"
print(text.count("a"))   # 3

# 1ï¸âƒ£1ï¸âƒ£ startswith() â†’ True/False return (start check kare)

text = "python programming"
print(text.startswith("python"))   # True

# 1ï¸âƒ£2ï¸âƒ£ endswith() â†’ True/False return (end check kare)

text = "hello world"
print(text.endswith("world"))   # True

# 1ï¸âƒ£3ï¸âƒ£ isdigit() â†’ Check karta hai string sirf numbers hai ya nahi

text = "12345"
print(text.isdigit())   # True

# 1ï¸âƒ£4ï¸âƒ£ isalpha() â†’ Check karta hai string sirf alphabets hai ya nahi

text = "Python"
print(text.isalpha())   # True

# 1ï¸âƒ£5ï¸âƒ£ isalnum() â†’ Alphabet + numbers allowed (space nahi)

text = "Python123"
print(text.isalnum())   # True

# â­ Bonus â€” String Slicing (Bahut Important)

text = "PYTHON"
print(text[0:3])   # PYT
print(text[:4])    # PYTH
print(text[-3:])   # HON
````

# ğŸ“ Python List

Python me List ek ordered, mutable (changeable) collection hoti hai, jisme multiple items (different data types) store kiye ja sakte hain.

`my_list = [10, "Hello", 3.5, True]`

````python
# 1ï¸âƒ£ append() â†’ List ke end me item add karta hai

nums = [1, 2, 3]
nums.append(4)
print(nums)   # [1, 2, 3, 4]

# 2ï¸âƒ£ insert(index, value) â†’ Specific position par value insert

nums = [10, 20, 30]
nums.insert(1, 15)
print(nums)   # [10, 15, 20, 30]

# 3ï¸âƒ£ extend() â†’ Ek list ke saath dusri list merge

a = [1, 2]
b = [3, 4]
a.extend(b)
print(a)   # [1, 2, 3, 4]

# 4ï¸âƒ£ remove() â†’ Value ko remove karta hai

nums = [1, 2, 3, 2]
nums.remove(2)
print(nums)   # [1, 3, 2]   (sirf first 2 remove hota hai)

# 5ï¸âƒ£ pop() â†’ Index ke base par element remove + return

nums = [10, 20, 30]
nums.pop(1)
print(nums)   # [10, 30]

# 6ï¸âƒ£ clear() â†’ Puri list empty

nums = [1, 2, 3]
nums.clear()
print(nums)   # []

# 7ï¸âƒ£ index() â†’ Value ka index return karta hai

nums = [5, 10, 15]
print(nums.index(10))   # 1

# 8ï¸âƒ£ count() â†’ Value kitni baar aayi hai

nums = [1, 2, 2, 3]
print(nums.count(2))   # 2

# 9ï¸âƒ£ sort() â†’ List ko ascending order me sort karta hai

nums = [4, 1, 3, 2]
nums.sort()
print(nums)   # [1, 2, 3, 4]

# ğŸ”Ÿ reverse() â†’ List ko ulta kar deta hai

nums = [1, 2, 3]
nums.reverse()
print(nums)   # [3, 2, 1]

# â­ Bonus Concept â€” List Slicing (Bahut Important)

nums = [10, 20, 30, 40, 50]
print(nums[1:4])   # [20, 30, 40]
print(nums[:3])    # [10, 20, 30]
print(nums[-2:])   # [40, 50]

# â­ List Loop (Iteration)

nums = [10, 20, 30]
for n in nums:
    print(n)
````

# ğŸ“ Python Tuple

Python me Tuple ek ordered, immutable (change-na-hone-wala) collection hota hai.
    
List jaisa hota hai, lekin tuple ko modify nahi kar sakte.

`my_tuple = (10, 20, 30, 40)`

* âœ” Indexing supported
* âœ” Duplicate values allowed
* âŒ Change / Add / Remove not allowed

````python
# 1ï¸âƒ£ count() â†’ Value kitni baar aayi hai

t = (1, 2, 2, 3)
print(t.count(2))   # 2

# 2ï¸âƒ£ index() â†’ Value ka index return

t = (10, 20, 30)
print(t.index(20))   # 1

# â­ Bonus â€” Tuple Slicing (List jaisa)

t = (10, 20, 30, 40, 50)
print(t[1:4])  # (20, 30, 40)
print(t[:3])   # (10, 20, 30)
print(t[-2:])  # (40, 50)


### ğŸ”¥ Tuple Packing & Unpacking (Bahut Important)
# ğŸ Packing

t = 10, 20, 30
print(t)   # (10, 20, 30)

# ğŸ“¦ Unpacking

a, b, c = (10, 20, 30)
print(a, b, c)   # 10 20 30

# Single Element Tuple Important Note

t = (10,)   # Single element tuple
print(type(t))   # <class 'tuple'>
````

````python
# 1ï¸âƒ£ How to Add Value in Tuple

### Tuple immutable hota hai, isliye direct add nahi kar sakte.
### But list me convert karke add kar sakte hain â†’ phir tuple me convert.

t = (10, 20, 30)
temp = list(t)
temp.append(40)
t = tuple(temp)
print(t)   # (10, 20, 30, 40)

# 2ï¸âƒ£ Tuple Slicing ===>  List jaisa hi slicing hota hai.

t = (10, 20, 30, 40, 50)
print(t[1:4])   # (20, 30, 40)
print(t[:3])    # (10, 20, 30)
print(t[-2:])   # (40, 50)

# 3ï¸âƒ£ Value Get (Indexing) ===>  Specific position ki value lene ke liye index use.

t = (5, 10, 15, 20)
print(t[2])   # 15
print(t[-1])  # 20

# 4ï¸âƒ£ Concatenation (2 Tuples Ko Jodna)   ===>>>  + operator se 2 tuples ko combine kar sakte hain.

t1 = (1, 2, 3)
t2 = (4, 5)
t3 = t1 + t2
print(t3)   # (1, 2, 3, 4, 5)

# 5ï¸âƒ£ Remove Value from Tuple

# Direct remove nahi hota, kyunki tuple immutable hai.
# But list ke through trick se remove kar sakte hain:

t = (10, 20, 30, 20)
temp = list(t)
temp.remove(20)  # first 20 remove
t = tuple(temp)
print(t)   # (10, 30, 20)

# 6ï¸âƒ£ del â€” Tuple Delete ===>> del se poora tuple delete ho jata hai, element delete nahi hota.

t = (10, 20, 30)
del t
print(t)   # âŒ Error â€” tuple does not exist

# 7ï¸âƒ£ join() â€” Important Clarification

# ğŸ‘‰ join() tuple ka method nahi hota
# join actually string method hota hai jo tuple ke strings ko join karta hai.

t = ("A", "B", "C")
result = "-".join(t)
print(result)   # A-B-C
````

# ğŸ“ Python Set

Python me Set ek unordered, unindexed, mutable collection hota hai jisme duplicate values allowed nahi hoti.

`my_set = {10, 20, 30, 40}`

````python
# 1ï¸âƒ£ add() â€” Set me new value add karna

s = {1, 2, 3}
s.add(4)
print(s)   # {1, 2, 3, 4}

# 2ï¸âƒ£ update() â€” Multiple values add / merge

s = {1, 2}
s.update([3, 4, 5])
print(s)   # {1, 2, 3, 4, 5}

# 3ï¸âƒ£ remove() â€” Value remove but nahi mili to error

s = {10, 20, 30}
s.remove(20)
print(s)   # {10, 30}

# 4ï¸âƒ£ discard() â€” Value remove karta par error nahi deta

s = {10, 20, 30}
s.discard(40)
print(s)   # {10, 20, 30}

# 5ï¸âƒ£ pop() â€” Random element remove + return

s = {1, 2, 3}
item = s.pop()
print(item)  # Random element
print(s)

# 6ï¸âƒ£ clear() â€” Set empty

s = {1, 2, 3}
s.clear()
print(s)   # set()

# âœ” union() â†’ Sare unique values (OR operation)

a = {1, 2, 3}
b = {3, 4, 5}
print(a.union(b))   # {1, 2, 3, 4, 5}

# âœ” intersection() â†’ Common values (AND operation)

a = {1, 2, 3}
b = {2, 3, 4}
print(a.intersection(b))   # {2, 3}

# âœ” difference() â†’ a me jo hain but b me nahi

a = {1, 2, 3}
b = {2, 3}
print(a.difference(b))   # {1}

# âœ” symmetric_difference() â†’ Unique in both (XOR)

a = {1, 2, 3}
b = {3, 4}
print(a.symmetric_difference(b))   # {1, 2, 4}

# ğŸ Check Membership â€” in keyword

s = {10, 20, 30}
print(20 in s)   # True
print(50 in s)   # False
````

# ğŸ“ Python Dictionary

Python me Dictionary ek unordered, mutable collection hota hai jo keyâ€“value pairs ke form me data store karta hai.

`student = {"name": "Rahul", "age": 20, "city": "Delhi"}`

````python
# 1ï¸âƒ£ get() â†’ Key ki value safely return karta hai

d = {"name": "Aman", "age": 22}
print(d.get("age"))         # 22
print(d.get("city"))        # None (Error nahi deta)

# 2ï¸âƒ£ keys() â†’ Saare keys return

d = {"a": 1, "b": 2}
print(d.keys())    # dict_keys(['a', 'b'])

# 3ï¸âƒ£ values() â†’ Saari values return

d = {"a": 1, "b": 2}
print(d.values())   # dict_values([1, 2])

# 4ï¸âƒ£ items() â†’ Key + Value pair return

d = {"a": 1, "b": 2}
print(d.items())    # dict_items([('a', 1), ('b', 2)])

# 5ï¸âƒ£ update() â†’ Dictionary me value update / add

d = {"a": 1, "b": 2}
d.update({"b": 20, "c": 30})
print(d)    # {'a': 1, 'b': 20, 'c': 30}

# 6ï¸âƒ£ pop() â†’ Specific key remove + value return

d = {"a": 1, "b": 2, "c": 3}
d.pop("b")
print(d)    # {'a': 1, 'c': 3}

# 7ï¸âƒ£ popitem() â†’ Last inserted keyâ€“value hatata + return

d = {"a": 1, "b": 2, "c": 3}
d.popitem()
print(d)    # {'a': 1, 'b': 2}

# 8ï¸âƒ£ clear() â†’ Puri dictionary empty

d = {"x": 10, "y": 20}
d.clear()
print(d)   # {}

# 9ï¸âƒ£ fromkeys() â†’ Keys ke base par dict create

keys = ["a", "b", "c"]
d = dict.fromkeys(keys, 0)
print(d)   # {'a': 0, 'b': 0, 'c': 0}

#ğŸ”¥ Dictionary â€” Add / Update / Remove Shortcut

#âœ” Add

d = {"x": 1}
d["y"] = 2
print(d)   # {'x': 1, 'y': 2}

# âœ” Update value

d["x"] = 100
print(d)   # {'x': 100, 'y': 2}

# âœ” Delete key

del d["y"]
print(d)   # {'x': 100}

# â­ Loop Dictionary

d = {"name": "Rohan", "age": 21}

for key in d:
    print(key, d[key])
````

# Argument vs Parameter Python

## ğŸ”· Parameter (Function Definition ke Time)

ğŸ“Œ Parameter wo variable hote hain **jo function define karte waqt likhe jaate hain**. 

Ye input receive karne ke placeholder hote hain.

ğŸ”¹ Parameter = Function ke andar data receive karne ke liye naam

### Example
```python
def add(a, b):   # a and b = parameters
    return a + b
```
## ğŸ”· Argument (Function Call ke Time)

ğŸ“Œ Argument wo real values hoti hain jo function ko call karte waqt pass ki jati hain.

```python
result = add(5, 10)   # 5 and 10 = arguments


| Feature          | Parameter                  | Argument                 |
| ---------------- | -------------------------- | ------------------------ |
| Kab use hota hai | Function define karte waqt | Function call karte waqt |
| Kya hota hai     | Variable name              | Actual value             |
| Position         | Function header            | Function calling         |
| Example          | `a, b`                     | `5, 10`                  |


`Parameter` = Function ke andar input receive karne ka placeholder,

`Argument` = Function ko call karte waqt diya gaya actual data
